I have been getting this question a lot lately:  "How can I make a blog similar
to yours?"  For those that want to jump ahead and try it before reading this
post check out the blog code
[repository](https://github.com/ocefpaf/python4oceanographers) and start

### Blog environment

I like to separate every new work environment in a different
[virtualenv](http://www.virtualenv.org/en/latest/).  And to manage my
virtualenvs I use [virtualenvwrapper](http://virtualenvwrapper.readthedocs.org/en/latest/)

You can install both using your distro package manager (or with any other python
installer out there: easy_install, pip, conda etc).  In my OpenSUSE box I just
use zypper,

    sudo zypper in python-virtualenvwrapper

and all the dependencies will be installed as well.

The `virtualenvwrapper` extends `virtualenv` with the several new commands, but
I usually use just these three:

* `mkvirtualenv` to create a new virtualenv.
* `workon` to enter the virtualenv.
* `deactivate` to exit the current virtualenv.

To crate the blog environment type:

    mkvirtualenv --distribute --no-site-packages blog

Then enter the environment with,

    workon blog

and finally download the file
[requirements.txt](https://raw.github.com/ocefpaf/python4oceanographers/master/content/downloads/code/requirements.txt)
and install all the modules that are needed with pip:

    pip install -r requirements.txt

Note that you do not need to be root to install inside the virtualenv.
You just need to be inside the virtualenv.
Check if everything is installed correctly by typing `pip freeze`
inside your virtualenv and compare with the original requirements.txt
The blog environment is almost ready.  There are still two other dependencies
to download the [theme](https://github.com/jakevdp/pelican-octopress-theme.git)
and the [plug-ins](https://github.com/getpelican/pelican-plugins.git).

* git clone https://github.com/jakevdp/pelican-octopress-theme.git
* git clone https://github.com/getpelican/pelican-plugins.git

There are several other themes and plug-ins for pelican.  I suggest you
check them out, but be aware that you will have to adapt the IPython notebook
plug-in inside the theme.  More information on how to do this in
[Jake's blog](http://jakevdp.github.io/blog/2013/05/07/migrating-from-octopress-to-pelican).

### The blog

Now there are two options to proceed:

#. clone my current blog and start from there,
#. or to start a new one using [pelican-quickstart](http://docs.getpelican.com/en/3.1.1/getting_started.html)

For brevity I will clone my current blog.

    git clone https://github.com/ocefpaf/python4oceanographers

If you choose to start from scratch I recommend taking a look at
both `publishconf.py` and `pelicanconf.py` from my repository.
However, if you choose to use my blog as a template,
change into the python4oceanographers directory, modify `pelicanconf.py` `path`
variable to point to where you cloned `pelican-octopress-theme` and `pelican-plugins`.

```
path = '/home/filipe/BLOG'
THEME = '%s/pelican-octopress-theme' % path
PLUGIN_PATH = '%s/pelican-plugins' % path
```

### Compiling the blog
Now you can use `make clean` to get a clean slate, and
`make html` to create the HTML output folder.

If everything is OK enter the `output` folder and type
`python -m SimpleHTTPServer 8080`.  Then open your browser at
[localhost](http://localhost:8080/) and check the results.

To use `make deploy` you must first edit Makefile to point to your repository.

### The blog work flow

My posts are just IPython notebooks processed into HTML files using
[Pelican](http://blog.getpelican.com/) and the
[liquid tags plug-in](https://github.com/jakevdp/pelican-plugins/blob/liquid_tags/liquid_tags/notebook.py)

There are other similar techniques to blog with IPython notebooks out there
like,

* http://www.damian.oquanta.info/
* http://blog.fperez.org/2012/09/blogging-with-ipython-notebook.html
* http://danielfrg.github.io/blog/2013/03/08/pelican-ipython-notebook-plugin/

but I'm following [Jake's technique](http://jakevdp.github.io) that pushes most
of the work to nbconvert making it possible it write all posts as `.ipynb`.

Now start your own blog!  Create a
[github repository for your page](http://pages.github.com/), adapt the
configuration to your case and Good luck!
